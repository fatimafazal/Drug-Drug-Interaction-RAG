```python
import json
import chromadb
```

```python
with open("prompt.json") as json_file:
    data = json.load(json_file)
```

```python
data=data[0:500]
```

```python
data[0]
```

```python
cache_dir = "D://cache" 
```

```python

documents = [item['prompt'] for item in data]
ids = [str(i) for i in range(len(documents))]
metadatas = [{k: v for k, v in item.items() if k != "prompt"} for item in data]
```

```python
documents[0]
```

```python
chroma_client = chromadb.Client()

collection = chroma_client.create_collection(
    metadata={"hnsw:space": "cosine"}, name="syllabi")
```

```python
collection.add(documents=documents, ids=ids, metadatas=metadatas)
```

```python
def pretty_print(content):
    print(json.dumps(content, indent=2))
```

```python
collection
```

```python
collection.get(ids=["1"])
```

```python
#
# Query by id.
#
# By default, embeddings are not returned:
pretty_print(collection.get(ids=["1"]))
pretty_print(collection.get(ids=["1"], include=["documents", "embeddings"]))
```

```python
# Filter by document content:
#pretty_print(collection.get(where_document={"$contains": "Bivalirudin"}))
# Limit results:
pretty_print(collection.get(limit=1, where_document={"$contains": "Ceritinib"}))
```

```python
#
# Query by similarity.
#
query_text = "Ceritinib"
n_results = 3
pretty_print(collection.query(n_results=n_results, query_texts=[query_text]))
```
